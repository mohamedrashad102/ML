```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(100, size=200)
y = np.random.randint(100, size=200)

markers = ['.', ',', 'o', 'v', '^', '<', '>', 's', 'p', 'h', '+', 'x', 'D', 'd', '|', '_']

plt.scatter(x, y, c='green', marker='*', s=75, alpha=1)
plt.show()
```

```python
years = [2000 + x for x in range(15)]
weights = np.random.randint(75, 90, size=15).tolist()

styles = ['-', '--', '-.', ':']

plt.plot(years,weights, c='g', lw=4 , linestyle = '--')
plt.show()
```

```python
x = ['c', 'c++', 'java', 'python']
y = [35, 70 , 26, 55]

plt.bar(x,y, color="green" , align='center' , width=0.4, edgecolor= 'blue', lw=2)
plt.show()
```

```python
ages = np.random.normal(20,1.5, 1000)


plt.hist(ages)
plt.show()
```

```python
langs = ['c', 'c++', 'java', 'python']
votes = [25, 27, 23, 25]

plt.pie(votes, labels=langs, )
plt.show()
```

```python
heights = np.random.normal(5, 10, 20)
plt.boxplot(heights)
plt.show()
```

```python
years = [2000 + x for x in range(15)]
incomes = np.random.randint(75, 90, size=15).tolist()
incomes_ticks = list(range(75,90))

plt.plot(years,incomes)
plt.title('Mohamed\'s Income (EG)', fontsize= 20)
plt.xlabel('Years')
plt.ylabel('Income')
plt.yticks(incomes_ticks, [ f'{x}k EG' for x in incomes_ticks]  )
plt.show()
```
